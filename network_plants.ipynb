```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
```

```python

# エクセルデータを読み込む
df = pd.read_excel('network_data.xlsx',sheet_name = "Edges")

df.columns
# ネットワークを作成
G = nx.from_pandas_edgelist(df, 'node1', 'node2', edge_attr='label')

weight=10
weight2=1


G.add_edge('上郷エンジン', '上郷デジタル', weight=weight)
G.add_edge('上郷保全', '上郷デジタル', weight=weight)
G.add_edge('上郷鋳造', '上郷デジタル', weight=weight)
G.add_edge('上郷鋳造', '上郷保全', weight=weight)
G.add_edge('上郷エンジン', '上郷保全', weight=weight)
G.add_edge('上郷エンジン', '上郷鋳造', weight=weight)

G.add_edge('上下品管', '上郷デジタル', weight=weight)
G.add_edge('下山工場', '下山FC', weight=weight)

G.add_edge('田原工場', '田原品管', weight=weight)
G.add_edge('田原エンジン', '田原品管', weight=weight)
G.add_edge('田原工場', '田原エンジン', weight=weight)


G.add_edge('三好機械', '三好品管', weight=weight)
G.add_edge('本社機械', '本社FC', weight=weight)

G.add_edge('明知デフ', '明知鋳造', weight=weight)
G.add_edge('明知デフ', '明知工場', weight=weight)
G.add_edge('明知鋳造', '明知工場', weight=weight)

G.add_edge('高岡塗装', '堤工場', weight=weight)

G.add_edge("CF", "技術", weight=weight2)
G.add_edge("CF", "物流", weight=weight2)
G.add_edge("CF", "生管", weight=weight2)
G.add_edge("CF", "モノエ", weight=weight2)
G.add_edge("技術", "物流", weight=weight2)
G.add_edge("技術", "生管", weight=weight2)
G.add_edge("技術", "モノエ", weight=weight2)
G.add_edge("物流", "生管", weight=weight2)
G.add_edge("物流", "モノエ", weight=weight2)
G.add_edge("生管", "モノエ", weight=weight2)

combinations = {
    "上郷": "lightblue",
    "上下": "lightblue",
    "下山": "lightblue",
    "三好": "lightgreen",
    "田原": "lightcoral",
    "衣浦": "lightpink",
    "明知": "lightgray",
    "高岡": "lightyellow",
    "堤": "lightyellow",
    "CF":"lightskyblue",
    "技術":"lightskyblue",
    "物流":"lightskyblue",
    "生管":"lightskyblue",
    "モノエ":"lightskyblue",
    "本社":"lightsalmon",
    
    
}    

# 各ノードの色をデフォルト（ここでは灰色）に設定
node_colors = {node: 'gray' for node in G.nodes}

# ノード名に特定の文字列が含まれる場合、そのノードの色を設定
for node in G.nodes:
    for text, color in combinations.items():
        if text in node:
            node_colors[node] = color

# ノードの色をリストに変換（描画順に対応）
node_color_list = [node_colors[node] for node in G.nodes]

# ノードを描画
# nx.draw_networkx_nodes(G, pos, node_color=node_color_list)


# ノードとエッジの位置を保存するために layout を作成
plt.figure(figsize=(25, 25))
pos = nx.spring_layout(G, k=2, iterations=150)

# ノードを描画
nx.draw_networkx_nodes(G, pos, node_color=node_color_list, edgecolors='black',linewidths=2,node_size=3300)

# エッジを描画
nx.draw_networkx_edges(G, pos)

# エッジのラベルを描画
edge_labels = nx.get_edge_attributes(G, 'label')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_family='MS Gothic')

# ノードのラベルを描画
nx.draw_networkx_labels(G, pos, font_size=10, font_color='black', font_family='MS Gothic')

with PdfPages('graph_output.pdf') as pdf:
    pdf.savefig()

plt.show()
```
